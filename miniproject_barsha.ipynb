```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
from PIL import Image, ImageTk
from ttkbootstrap.style import Style
from ttkbootstrap import Meter
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import random
import matplotlib.pyplot as plt
import datetime

root = tk.Tk()
root.title("Mini Project")
root.geometry("800x400")
root.minsize(500, 400)

style = Style()
mainFrame = tk.Frame(root)
mainFrame.pack(fill=tk.BOTH, expand=True)

header = tk.Frame(mainFrame, borderwidth=5)
header.pack(side=tk.TOP, fill="x")

titleFrame = tk.Frame(header)
titleFrame.pack(side=tk.LEFT)

btnFrame = tk.Frame(header)
btnFrame.pack(side=tk.RIGHT)

contentFrame = tk.Frame(mainFrame)
contentFrame.pack(fill=tk.BOTH, expand=True)
contentFrame.grid_rowconfigure(0, weight=1)
contentFrame.grid_columnconfigure(0, weight=1)

# Home Page
global bckImg
homePage = tk.Frame(contentFrame)
homePage.grid(row=0, column=0, sticky="nsew")
imgFrame = tk.Frame(homePage)
imgFrame.pack(anchor="nw", side=tk.TOP, fill=tk.BOTH, expand=True, pady=30, padx=20)

bckImg = Image.open("images.jpg").resize((500, 400))
photu = ImageTk.PhotoImage(image=bckImg)
ImgLabel = tk.Label(imgFrame, image=photu, pady=15)
ImgLabel.pack(fill=tk.BOTH, expand=True)

textContent = '''Welcome to the Soil Monitoring System!
This system helps monitor soil moisture, temperature, conductivity, Nitrogen, potassiun ,phosphorus and pH levels
to support smart farming and efficient irrigation. Can predict suitable crop based on these parameters.
It will also help farmers in early detection of crop diseases. 
'''

aboutLabel = tk.Label(imgFrame, text="About", font="Arial 18 bold")
aboutLabel.pack(fill="x", padx=5, pady=15, side=tk.TOP)

text_Label = tk.Label(imgFrame, text=textContent, font="Arial 12 bold")
text_Label.pack(fill="x", padx=5)
#Home Page Khatam


# Service Page
def OpenDialog():
    messagebox.showinfo("Info", "This feature will work in next version using machine learning.")

servicePage = tk.Frame(contentFrame)
servicePage.grid(row=0, column=0, sticky="nsew")

mainServiceFrame = tk.Frame(servicePage)
mainServiceFrame.pack(fill=tk.BOTH, expand=True)

serviceFrame1 = tk.Frame(mainServiceFrame, bg="#a89191")
serviceFrame1.pack(side=tk.LEFT, expand=True, anchor='w', padx=8, pady=15)
global CropDetectionImg
CropDetectionImg = Image.open("images2.jpg").resize((400, 350))
photu2 = ImageTk.PhotoImage(image=CropDetectionImg)
DiseaseImgLabel = tk.Label(serviceFrame1, image=photu2)
DiseaseImgLabel.pack(fill=tk.BOTH, expand=True, pady=20, padx=10)
label1Button = tk.Button(serviceFrame1, text="Predict Crop Disease", command=OpenDialog)
label1Button.pack(side=tk.BOTTOM, anchor="center")

serviceFrame2 = tk.Frame(mainServiceFrame, bg="#a89191")
serviceFrame2.pack(expand=True, padx=8, pady=15, anchor="e", side=tk.RIGHT)
global CropPredictionImg
CropPredictionImg = Image.open("images3.jpg").resize((400, 350))
photu3 = ImageTk.PhotoImage(image=CropPredictionImg)
CropImgLabel = tk.Label(serviceFrame2, image=photu3)
CropImgLabel.pack(fill=tk.BOTH, expand=True, pady=20, padx=10)
label2Button = tk.Button(serviceFrame2, text="Predict suitable crop", command=OpenDialog)
label2Button.pack(side=tk.BOTTOM, anchor="center")
# Service Page Kahtam


# Sensor Page
nitrogendata = []
Phosphorusdata = []
Potassiumdata = []
Temperaturedata = []
pHdata = []
Humiditydata = []
timeData = []

sensorPage = tk.Frame(contentFrame, bg="beige")
sensorPage.grid(row=0, column=0, sticky="nsew", pady=10)
contentFrame.grid_rowconfigure(0, weight=1)
contentFrame.grid_columnconfigure(0, weight=1)

SensorLabel = tk.Label(sensorPage, text="Real Time Sensor data", font="Arial 20 bold", padx=5, pady=5)
SensorLabel.pack(side=tk.TOP, pady=35)
tableFrame = tk.Frame(sensorPage)
tableFrame.pack(fill=tk.BOTH)

column = ("Nitrogen (N)", "Phosphorus (P)", "Potassium (K)", "Temperature C", "pH", "Humidity")
table = ttk.Treeview(tableFrame, columns=column, show="headings", height=1)
table.pack(fill=tk.BOTH)

for col in column:
    table.heading(col, text=col, anchor="center")
    table.column(col, anchor="center")

graphLabel = tk.Label(sensorPage, text="Graphs", padx=5, pady=2, font="Arial 20 bold")
graphLabel.pack(padx=5, pady=10, anchor="center")

meterFrame = tk.Frame(sensorPage)
meterFrame.pack(pady=20)

meters = {}
for col in column:
    gauge = tk.Frame(meterFrame, bg="#F5F5DC")
    gauge.pack(side=tk.LEFT, padx=10)
    meter = Meter(gauge, metersize=150, padding=10, amounttotal=100, amountused=0,
                  metertype="semi", subtext=col, interactive=False, bootstyle="info")
    meter.pack()
    meters[col] = meter

def table_data():
    table.delete(*table.get_children())
    data = (random.randint(10, 150), random.randint(15, 100), random.randint(15, 100),
            round(random.uniform(15, 40), 2), round(random.uniform(5, 8), 2),
            round(random.uniform(30, 95), 2))
    listdata = list(data)
    nitrogendata.append(listdata[0])
    Phosphorusdata.append(listdata[1])
    Potassiumdata.append(listdata[2])
    Temperaturedata.append(listdata[3])
    pHdata.append(listdata[4])
    Humiditydata.append(listdata[5])
    table.insert("", "end", values=data)

    timeData.append(datetime.datetime.now().strftime("%H:%M:%S"))

    if len(timeData) > 5:
        timeData.pop(0)
        nitrogendata.pop(0)
        Phosphorusdata.pop(0)
        Potassiumdata.pop(0)
        Temperaturedata.pop(0)
        pHdata.pop(0)
        Humiditydata.pop(0)

    for i, col in enumerate(column):
        meters[col].configure(amountused=listdata[i])

    update_alerts()
    update_graph()

    root.after(3000, table_data)

# Sensor Khatam

# Scrollable Alerts Page
alertsPage = tk.Frame(contentFrame)
alertsPage.grid(row=0, column=0, sticky="nsew")

canvas = tk.Canvas(alertsPage)
scrollbar = ttk.Scrollbar(alertsPage,orient="vertical", command=canvas.yview)
scrollable_frame = tk.Frame(canvas)

scrollable_frame.bind("<Configure>",lambda e: canvas.configure(scrollregion=canvas.bbox("all")))
canvas.create_window((0, 0), window=scrollable_frame, anchor="nw")
canvas.configure(yscrollcommand=scrollbar.set)

canvas.pack(side="left", fill="both", expand=True)
scrollbar.pack(side="right",fill="y")

alertTitle = tk.Label(scrollable_frame, text="Alerts", font=("Arial", 18, "bold"))
alertTitle.pack(anchor="center",pady=10)

alertLabel = tk.Label(scrollable_frame, text="Soil Health Results", font=("Arial", 12))
alertLabel.pack(anchor="center",pady=5)

graphFrame = tk.Frame(scrollable_frame)
graphFrame.pack(fill=tk.BOTH, expand=True, pady=10)

def update_graph():
    for widget in graphFrame.winfo_children():
        widget.destroy()  # Clear old graphs

    trendLabel = tk.Label(graphFrame, text="History / Trends",font=("Arial", 18, "bold"))
    trendLabel.pack(anchor="center", pady=5)

    fig, axes = plt.subplots(2, 3, figsize=(9, 5), dpi=100)
    fig.subplots_adjust(hspace=1.3,wspace=0.5)

    datasets = [("Nitrogen (N)", nitrogendata),("Phosphorus (P)", Phosphorusdata),("Potassium (K)", Potassiumdata),("Temperature (¬∞C)",Temperaturedata),
                ("pH", pHdata),("Humidity (%)", Humiditydata)]

    for ax, (label, data) in zip(axes.flat, datasets):
        if len(timeData) > 0:
            ax.plot(timeData[-5:], data[-5:], marker='o', label=label)
            ax.set_title(label)
            ax.set_xlabel("Time")
            ax.set_ylabel("Value")
            ax.grid(True, linestyle="--", alpha=0.6)
            plt.setp(ax.get_xticklabels(), rotation=45, ha="right")

    canvas_plot = FigureCanvasTkAgg(fig, master=graphFrame)
    canvas_plot.draw()
    canvas_plot.get_tk_widget().pack(fill=tk.BOTH, expand=True, pady=6)

def update_alerts():
    msg = []
    if len(nitrogendata) > 0:
        N = nitrogendata[-1]
        P = Phosphorusdata[-1]
        K = Potassiumdata[-1]
        T = Temperaturedata[-1]
        pH_val = pHdata[-1]
        H = Humiditydata[-1]

        if N < 20:
            msg.append("‚ö†Ô∏è Nitrogen level too low ‚Äî add nitrogen-based fertilizer.")
        elif N > 120:
            msg.append("üö´ Nitrogen level too high ‚Äî may cause burning of roots.")

        if P < 20:
            msg.append("‚ö†Ô∏è Phosphorus level low ‚Äî consider DAP or compost.")
        elif P > 90:
            msg.append("üö´ Excess Phosphorus ‚Äî can hinder micronutrient uptake.")

        if K < 20:
            msg.append("‚ö†Ô∏è Potassium deficiency detected.")
        elif K > 90:
            msg.append("üö´ Too much Potassium ‚Äî can affect calcium and magnesium absorption.")

        if T < 15:
            msg.append(f"‚ùÑÔ∏è Temperature too low ({T}¬∞C). Risk of stunted growth.")
        elif T > 35:
            msg.append(f"üå°Ô∏è High temperature ({T}¬∞C). Irrigate to cool soil.")

        if pH_val < 5.5:
            msg.append(f"‚ö†Ô∏è Soil too acidic (pH {pH_val}). Add lime or ash.")
        elif pH_val > 8:
            msg.append(f"‚ö†Ô∏è Soil too alkaline (pH {pH_val}). Add organic compost.")

        if H < 40:
            msg.append(f"üíß Low humidity ({H}%). Increase irrigation.")
        elif H > 90:
            msg.append(f"üí¶ Excess humidity ({H}%) may promote fungal diseases.")

        if not msg:
            msg = ["Not to Worry!!! All soil parameters are within safe range."]
        alertLabel.config(text="\n".join(msg))

table_data()
update_graph()
update_alerts()
# Aletrs Khatam

# Changing Pages
def show_pages(page):
    page.tkraise()

titleLabel = tk.Label(titleFrame, text="Soil Monitoring System", font=("Comic Sans MS", 16, "bold"), padx=6, pady=4)
titleLabel.pack(padx=100, pady=12, anchor="w")

Alert_btn = tk.Button(btnFrame, borderwidth=3, text="Alerts", command=lambda: show_pages(alertsPage))
Alert_btn.pack(side=tk.RIGHT, anchor="ne", padx=8)

Sens_btn = tk.Button(btnFrame, borderwidth=3, text="Sensor Data", command=lambda: show_pages(sensorPage))
Sens_btn.pack(side=tk.RIGHT, anchor="ne", padx=8)

Serv_btn = tk.Button(btnFrame, borderwidth=3, text="Services", command=lambda: show_pages(servicePage))
Serv_btn.pack(side=tk.RIGHT, anchor="ne", padx=8)

home_btn = tk.Button(btnFrame, borderwidth=3, text="HOME", command=lambda: show_pages(homePage))
home_btn.pack(side=tk.RIGHT, anchor="ne", padx=8)

homePage.tkraise()
root.mainloop()
```
